```python
class Perceptron:
    def __init__(self, input_size):#constructeur
        # Initialisation alatoire des poids et du biais
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        print(" les poids sont:" )
        print(  self.weights)
        print(" le biais est :")
        print( self.bias)

    def activation_function(self, x):
        # Fonction d'activation (fonction de Heaviside)
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        # Calcul de la somme pondre
        print( "la fonction dot me permets de faire la multiplication entre weights et inputs ")
        print(inputs )
        print(self.weights)
        print(self.bias)
        sum = np.dot(inputs, self.weights) + self.bias # somme des Wi Xi  + bias *1
        # Application de la fonction d'activation
        return self.activation_function(sum)
```

```python

def create_perceptron_graph(perceptron):
    # Crer un objet Digraph
    dot = Digraph(comment='Perceptron')
    dot.attr(rankdir='LR')  # Orientation de gauche  droite

    # Dfinir les styles
    dot.attr('node', shape='circle')

    # Ajouter les nuds d'entre
    for i in range(1, len(perceptron.weights)):
        dot.node(f'x{i}', f'x{i}')

    # Ajouter le nud de sortie
    dot.node('y', 'y')

    # Ajouter le nud de biais
    dot.node('b', f'Biais\n{perceptron.weights[0]:.4f}', shape='box')

    # Ajouter les connexions avec les poids
    for i in range(1, len(perceptron.weights)):
        dot.edge(f'x{i}', 'y', label=f'{perceptron.weights[i]:.4f}')
    dot.edge('b', 'y')

    # Ajouter une note pour la fonction d'activation
    dot.node('activation', 'Fonction d\'activation:\nSeuil  0', shape='note')
    dot.edge('y', 'activation', style='dashed', arrowhead='none')

    # Ajouter des informations sur l'apprentissage
    dot.node('info', f'Epochs: {perceptron.nb_epochs}\nLearning Rate: {perceptron.learning_rate}', shape='note')

    # Rendre le graphe
    dot.render('perceptron', format='png', view=True)
```

==================================================apprentissage des neurones===============================

```python

class Perceptron(object):

    def __init__(self, no_of_inputs, nb_epochs=3, learning_rate=0.001):
        self.nb_epochs = nb_epochs
        self.learning_rate = learning_rate #pas d'appentissage
        self.weights = np.zeros(no_of_inputs + 1)#nb inputs + bias
        print("les poids sont:" ,self.weights)
        print("le bias est:" ,self.weights[0])

    def affiche_weights(self):
        print("les poids sont:" ,self.weights[1:])
        print("le bias est:" ,self.weights[0])
        print("le vecteur W:" ,self.weights)

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0] # W0*1 +   w1*X1 + w2*X2  +....
        if summation > 0:
            activation = 1
        else:
            activation = 0
        return activation

    def train(self, training_inputs, labels):
        for e in range(self.nb_epochs): # boucler 100 fois
            print( "epoch numro : ",e )
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)# rsultat prdit par mon neurone
                # apres chaque prdiction  je vais modifier les poids et le bias
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs  #label = rsultat que je devrais avoir
                print( "ecart =",label - prediction)
                self.weights[0] += self.learning_rate * (label - prediction)* 1#modif du poids du bias

                print("je modifie les poids et le bias")
                print(self.weights)
```

```python
from graphviz import Digraph
from sklearn.neural_network import MLPClassifier
import numpy as np

def visualiser_mlp(mlp):
    dot = Digraph(comment='MLP')
    dot.attr(rankdir='LR', size='8,5')

    # Couche d'entre
    dot.attr('node', shape='circle', color='lightblue', style='filled')
    for i in range(mlp.n_features_in_):
        dot.node(f'I{i}', f'I{i}')

    # Couches caches
    dot.attr('node', shape='circle', color='lightgreen', style='filled')
    for l, taille_couche in enumerate(mlp.hidden_layer_sizes):
        for i in range(taille_couche):
            dot.node(f'H{l+1}_{i}', f'H{l+1}_{i}')

    # Couche de sortie
    dot.attr('node', shape='circle', color='lightsalmon', style='filled')
    for i in range(mlp.n_outputs_):
        dot.node(f'O{i}', f'O{i}')

    # Connexions
    dot.attr('edge', color='gray')
    # De l'entre  la premire couche cache
    for i in range(mlp.n_features_in_):
        for j in range(mlp.hidden_layer_sizes[0]):
            dot.edge(f'I{i}', f'H1_{j}')

    # Entre les couches caches
    for l in range(len(mlp.hidden_layer_sizes) - 1):
        for i in range(mlp.hidden_layer_sizes[l]):
            for j in range(mlp.hidden_layer_sizes[l+1]):
                dot.edge(f'H{l+1}_{i}', f'H{l+2}_{j}')

    # De la dernire couche cache  la sortie
    for i in range(mlp.hidden_layer_sizes[-1]):
        for j in range(mlp.n_outputs_):
            dot.edge(f'H{len(mlp.hidden_layer_sizes)}_{i}', f'O{j}')

    return dot
```

```python
# fonction uniquement pour l'affichage du reseau de neurones
from graphviz import Digraph

def plot_nn_graphviz(layer_sizes):
    dot = Digraph(comment='Neural Network')
    dot.attr(rankdir='LR', size='8,5')

    dot.attr('node', shape='circle')

    for i, layer_size in enumerate(layer_sizes):
        with dot.subgraph(name=f'cluster_{i}') as c:
            c.attr(color='lightgrey')
            c.attr(label=f'Layer {i}')
            for j in range(layer_size):
                c.node(f'L{i}N{j}', f'N{j}')

    # Connections
    for i in range(len(layer_sizes) - 1):
        for j in range(layer_sizes[i]):
            for k in range(layer_sizes[i+1]):
                dot.edge(f'L{i}N{j}', f'L{i+1}N{k}')

    dot.render('neural_network', view=True, format='png')

# Utilisation
layer_sizes = [4, 10, 5, 3]
plot_nn_graphviz(layer_sizes)
```
