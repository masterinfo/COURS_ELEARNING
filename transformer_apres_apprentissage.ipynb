```python

import torch
from transformers import MarianMTModel, MarianTokenizer

# MarianMTModel est un modèle de traduction automatique neuronale (NMT) basé sur l'architecture Transformer, conçu pour la traduction multilingue.

class TranslationModel:
    def __init__(self, model_name="Helsinki-NLP/opus-mt-en-fr"):
        #Helsinki-NLP/opus-mt-en-fr est un modèle de traduction automatique neuronale développé par l'équipe Helsinki-NLP pour traduire du texte de l'anglais vers le français
        self.model = MarianMTModel.from_pretrained(model_name)
        self.tokenizer = MarianTokenizer.from_pretrained(model_name)
        #Collection de textes traduits provenant du web Couvre plus de 40 langues avec des paires de phrases alignées Contient environ 55 millions de paires de phrases

    def translate(self, source_sentence, max_length=50):
        # Encoder la phrase source
        encoder_input = self.tokenizer.encode(source_sentence, return_tensors="pt")
        #Il indique au tokenizer de renvoyer les résultats sous forme de tenseurs PyTorch Sans cette option, le tokenizer renverrait généralement des listes Python standard7


        # Générer la traduction
        translated = self.model.generate(encoder_input, max_length=max_length)

        # Décoder la sortie
        translated_sentence = self.tokenizer.decode(translated[0], skip_special_tokens=True)

        return translated_sentence


def translate_simple(model, source_sentence, max_length=50):
    return model.translate(source_sentence, max_length)


# Utilisation
model = TranslationModel()
source_sentence = "Hello world, how are you ?"
translation = translate_simple(model, source_sentence)
print(f"Source: {source_sentence}")
print(f"Traduction: {translation}")
```
