```python
# prompt: affiche en 3 dimension les iris avec seaborn

import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'iris' is already loaded as in the provided code
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Choose three features for the 3D plot (you can experiment with different combinations)
x_axis = X[:, 0]  # Sepal length
y_axis = X[:, 1]  # Sepal width
z_axis = X[:, 2]  # Petal length


# Scatter plot with color-coded species
scatter = ax.scatter(x_axis, y_axis, z_axis, c=y, cmap='viridis')

# Customize the plot
ax.set_xlabel(feature_names[0])
ax.set_ylabel(feature_names[1])
ax.set_zlabel(feature_names[2])
ax.set_title('3D Scatter Plot of Iris Dataset')


# Add a colorbar to show the mapping of colors to species
plt.colorbar(scatter, label='Species')

plt.show()
```

Le diagramme d’Inselberg, aussi appelé Inselberg diagram ou parallel coordinates plot (coordonnées parallèles) dans la littérature anglophone, est une méthode classique pour visualiser des données multidimensionnelles en alignant plusieurs axes verticaux côte à côte, chaque observation étant représentée par une ligne reliant les valeurs sur chaque axe

```python
import pandas as pd
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le jeu de données Iris
data = sns.load_dataset('iris')

# Créer le diagramme de coordonnées parallèles
plt.figure(figsize=(10, 5))
parallel_coordinates(data, 'species', colormap=plt.get_cmap("Set2"))
plt.title('Diagramme de coordonnées parallèles (Inselberg) avec Iris')
plt.ylabel('Valeur normalisée')
plt.show()
```

```python
# prompt: foxplot sur les data iris

import matplotlib.pyplot as plt
# Display the foxplot
data = sns.load_dataset('iris')

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Boxplot of Iris Dataset Features')
plt.ylabel('Value')
plt.show()

# Boxplot by species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(data.columns[:-1]):
    plt.subplot(2, 2, i + 1) # 2 rows, 2 columns, current plot index
    sns.boxplot(x='species', y=feature, data=data)
    plt.title(f'Boxplot of {feature} by Species')
    plt.ylabel(feature)
    plt.xlabel('Species')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,   # pour la reproductibilité
    stratify=y         # optionnel mais bien pour garder les proportions de classes
)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
X_train[:10]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train[:10]
```

```python
from sklearn.linear_model import LogisticRegression# Train a logistic regression model on the training data
from sklearn.metrics import accuracy_score


# Entraînement du modèle
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Prédictions
y_pred = model.predict(X_test)

# Accuracy (attention à l’ordre: y_true, y_pred)
print(
    "The accuracy of the Logistic Regression is",
    accuracy_score(y_test, y_pred)
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

classifier = KNeighborsClassifier(n_neighbors=2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```
